```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
trip = pd.read_csv('trip.csv')
station = pd.read_csv('station.csv')
weather = pd.read_csv('weather.csv')
```

### Q1: Is there difference by weather conditions in the number of trips
(impossible weather filtering, boxplot)
(make weather condition features (rainy warm, rainy windy cold... : if elif else)
(pie chart of days distribution)

```python
weather['Precipitation_In '].replace('T','0', inplace=True)

weather['precipitation'] = weather['Precipitation_In '].astype(float)
```

```python
for c in ['Max_Temperature_F', 'Mean_Temperature_F', 'Min_TemperatureF',
       'Max_Dew_Point_F', 'MeanDew_Point_F', 'Min_Dewpoint_F', 'Max_Humidity',
       'Mean_Humidity ', 'Min_Humidity ', 'Max_Sea_Level_Pressure_In ',
       'Mean_Sea_Level_Pressure_In ', 'Min_Sea_Level_Pressure_In ',
       'Max_Visibility_Miles ', 'Mean_Visibility_Miles ',
       'Min_Visibility_Miles ', 'Max_Wind_Speed_MPH ', 'Mean_Wind_Speed_MPH ',
       'Max_Gust_Speed_MPH', 'precipitation', 'Cloud_Cover ']:
    weather.boxplot(c)
    plt.show()
```

```python
weather['Min_TemperatureF'] = np.where(weather['Min_TemperatureF']>40000,
                                       weather['Mean_Temperature_F'], weather['Min_TemperatureF'])

weather['Max_Temperature_F'] = np.where(weather['Max_Temperature_F']==0,
                                       weather['Mean_Temperature_F'], weather['Max_Temperature_F'])
```

```python
weather['cold'] =  np.where(weather.Min_TemperatureF<40,'cold',
                          np.where(weather['Max_Temperature_F']>70,'hot', 'normal'))
```

```python
weather.groupby('cold').size()
```

```python
weather.groupby('windy').size()
```

```python
weather['windy'] = np.where(weather['Max_Wind_Speed_MPH ']>25,'windy','not_windy')
```

```python
weather['Events'].replace('rain', 'Rain', inplace=True)
```

```python
weather['cat'] = weather.windy+'_'+weather.cold+'_'+weather.Events
```

```python
weather.groupby('cat').size().plot.pie()
```

```python
zipdict = np.load('zipcode_dict.npy').item()

df_rows = []
for k,v in zipdict.items():
    for vv in v:
        df_rows.append([k,vv])

zipdf = pd.DataFrame(df_rows, columns=['landmark','zipcode'])

zipdf['zip'] = zipdf['zipcode'].apply(lambda x: int(x.strip(',')))

zipdf.drop('zipcode', axis=1, inplace=True)
```

```python
station_z = pd.merge(station, zipdf, how='left', left_on='zip_code', right_on='zip')
```

```python
trip_z = pd.merge(trip, station_z[['name', 'landmark','zip_code']], how='left', left_on='Start Station', right_on='name')
```

```python
trip_w = pd.merge(trip_z, weather[['date_to_join','landmark','Max_Temperature_F','Mean_Temperature_F']], how='left', 
                  on=['date_to_join','landmark'])
```

```python
trip_w[['Start Date', 'End Date',
        'End Station', 'Subscription Type','Start Station','zip_code','landmark',
        'date_to_join','Max_Temperature_F','Mean_Temperature_F']].sort_values('date_to_join').head(10)
```

```python
trip_w = pd.merge(trip_z, weather[['date_to_join','landmark','cat']], how='left', 
                  on=['date_to_join','landmark'])
```

```python
trip_bycat = trip_w.groupby('cat').size().sort_values().reset_index()
trip_bycat['percent'] = trip_bycat[0]*100/len(trip_w)
trip_bycat.columns = ['cat','num','trip_percent']
```

```python
w_bycat = weather.groupby('cat').size().sort_values().reset_index()
w_bycat['percent'] = w_bycat[0]*100/len(weather)
w_bycat.columns = ['cat', 'num', 'weather_percent']
```

```python
bycat = pd.merge(trip_bycat[['cat','trip_percent']],
                w_bycat[['cat','weather_percent']])
```

```python
bycat.plot(kind='bar', x='cat')
```

### Q2: Is there a difference between Subscriber/Customer in daily routine?
(hour column, bar chart)
(weekend/weekday, bar chart extra)

```python
trip['start_hour'] = pd.to_datetime(trip['start_date']).dt.hour
```

```python
trip.groupby('start_hour').size().plot(kind='bar')
```

```python
bytypehour = trip.groupby(['Subscription Type','start_hour']).size().reset_index()
```

```python
bytypehour.pivot(index='start_hour', columns='Subscription Type', values=0).plot(kind='bar')
```

```python
trip['day_of_week'] = pd.to_datetime(trip['start_date']).dt.weekday_name

trip['weekend'] = trip['day_of_week'].isin(['Saturday', 'Sunday'])
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)#, sharey='none')
bytypehour_weekend = trip[trip['weekend']==True].groupby(['Subscription Type','start_hour']).size().reset_index()
bytypehour_weekend.pivot(index='start_hour', columns='Subscription Type', values=0).plot(kind='bar',
                                                                                         title='weekend',
                                                                                         ax=axes[0])
bytypehour_weekday = trip[trip['weekend']==False].groupby(['Subscription Type','start_hour']).size().reset_index()
bytypehour_weekday.pivot(index='start_hour', columns='Subscription Type', values=0).plot(kind='bar',
                                                                                         figsize=(14,6),
                                                                                         title='weekday',
                                                                                         ax=axes[1])

plt.show()
```

### Q3: Is there a difference between stations?
(avg duration, Customer ratio, bikenum, scatter plot)

```python
trip['duration'] = (pd.to_datetime(trip['end_date']) - pd.to_datetime(trip['start_date'])).astype('timedelta64[m]')
```

```python
bystation = trip.groupby(['Start Station']).agg({
    'duration':'mean',
    'Trip ID':'count'
}).reset_index()
```

```python
customer_ration = trip.groupby(['Start Station', 'Subscription Type']).size().reset_index()
```

```python
customer_ration = customer_ration.pivot(index='Start Station', columns='Subscription Type', values=0).reset_index()
```

```python
customer_ration['customer_ratio'] = customer_ration['Customer']/customer_ration['Subscriber']
```

```python
bystation = pd.merge(bystation, customer_ration[['Start Station', 'customer_ratio']], how='left')
```

```python
bystation = pd.merge(bystation, station[['name', 'dockcount']], how='left', left_on='Start Station', right_on='name')
```

```python
bystation.plot.scatter(x='duration', y='Trip ID', #s=bystation['dockcount'],
                      s=bystation['customer_ratio']*10, figsize=(10,7),
                      xlim=(0,150))
```

```python
bystation.plot.scatter(x='duration', y='customer_ratio', #s=bystation['dockcount'],
                      s=bystation['Trip ID']/20, figsize=(10,7))
```
