### Q2: Is there a difference between Subscriber/Customer in daily routine?
(hour column, bar chart)
(weekend/weekday, bar chart extra)

```python
trip['start_hour'] = pd.to_datetime(trip['start_date']).dt.hour
```

```python
trip.groupby('start_hour').size().plot(kind='bar')
```

```python
bytypehour = trip.groupby(['Subscription Type','start_hour']).size().reset_index()
```

```python
bytypehour.pivot(index='start_hour', columns='Subscription Type', values=0).plot(kind='bar')
```

```python
trip['day_of_week'] = pd.to_datetime(trip['start_date']).dt.weekday_name
```

```python
trip['weekend'] = trip['day_of_week'].isin(['Saturday', 'Sunday'])
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)#, sharey='none')
bytypehour_weekend = trip[trip['weekend']==True].groupby(['Subscription Type','start_hour']).size().reset_index()
bytypehour_weekend.pivot(index='start_hour', columns='Subscription Type', values=0).plot(kind='bar',
                                                                                         title='weekend',
                                                                                         ax=axes[0])
bytypehour_weekday = trip[trip['weekend']==False].groupby(['Subscription Type','start_hour']).size().reset_index()
bytypehour_weekday.pivot(index='start_hour', columns='Subscription Type', values=0).plot(kind='bar',
                                                                                         figsize=(14,6),
                                                                                         title='weekday',
                                                                                         ax=axes[1])

plt.show()
```
